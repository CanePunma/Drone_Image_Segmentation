```python
base_images = ["/home/ai2-jedi/Documents/code/Drone_Project/Seg_Data_Creation_Base_Images/reinforcement_1.png",
              "/home/ai2-jedi/Documents/code/Drone_Project/Seg_Data_Creation_Base_Images/reinforcement_2.png",
              "/home/ai2-jedi/Documents/code/Drone_Project/Seg_Data_Creation_Base_Images/reinforcement_3.png",
              "/home/ai2-jedi/Documents/code/Drone_Project/Seg_Data_Creation_Base_Images/reinforcement_4.png"]

penalization = 0.75
method = "K2"
con_color=True

print("Expect this to take "+str(len(sim_images_class)*3.66/3600)+" hours")

start = time.time()

count = 0
colored_clusters = []
failed = []
for image in sim_images_class:
    count += 1
    try:
        final_image = NIOSC3_5(base_images,image,penalization,method,caffe_objects,con_color)
        colored_clusters.append(final_image)
        if count%10==0 or count==1:
            print("Working on image #"+str(count)+", and the next 9 images!")
    except:
        print(str(count)+" didn't work!")
        failed.append(count-1)

end = time.time()

print("this took "+str((end-start)/3600)+" hours")
```
